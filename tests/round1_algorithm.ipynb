```python
df = pd.read_csv(prices_csv_file, sep=';')
```

```python
mp = df['mid_price'].values
```

```python
def moving_avg(v, ws):

    res = []

    cur = 0
    for i, num in enumerate(v):

        cur += num

        if i >= ws:
            cur -= v[i - ws] 
            res += [cur / ws]
        else:
            res += [None]
    
    return res
```

```python
moving_avg(mp, 10)
```

```python
plt.plot(mp)
plt.plot(moving_avg(mp, 10))
```
